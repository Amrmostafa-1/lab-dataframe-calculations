## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries
import numpy as np
import pandas as pd
```

Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# your code here
orders = pd.read_csv("Orders.csv")
orders.head()
```

We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1: How to aggregate the  amount_spent for unique customers?
#Group by customer ID and sum everything
#sort by amount spent

grouped_orders = orders.groupby('CustomerID')['amount_spent'].sum().reset_index()
sorted_grouped_orders = grouped_orders.sort_values(by = 'amount_spent', ascending= False)
sorted_grouped_orders
```

```python
#Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?
#Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

# Calculate the 95th and 75th percentiles
p75 = grouped_orders['amount_spent'].quantile(0.75)
p95 = grouped_orders['amount_spent'].quantile(0.95)



#Fileter and label the customers

VIP = grouped_orders[grouped_orders['amount_spent'] > p95]
Preferred = grouped_orders[(grouped_orders['amount_spent'] > p75) & (grouped_orders['amount_spent'] <= p95)]
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# Filter by VIP cutomers

vip_id = VIP['CustomerID'].tolist()
orders_vip = orders[orders['CustomerID'].isin(vip_id)]

#Do a count for the countries
orders_vip['Country'].value_counts()


#UK has the most VIP customers
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# Filter by VIP and preferred customers

vip_id = VIP['CustomerID'].tolist()
preferred_id = Preferred['CustomerID'].tolist()

# Combine both lists
combined_ids = vip_id + preferred_id
orders_v_p = orders[orders['CustomerID'].isin(combined_ids)]


#Do a count for the countries
orders_v_p['Country'].value_counts()

# UK has the most VIP and Preferred customers
```
